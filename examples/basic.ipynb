```python
import datetime as dt

import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
plt.rcParams["figure.figsize"] = (20,10)

from gsvi.connection import GoogleConnection
from gsvi.timeseries import SVSeries
from gsvi.catcodes import CategoryCodes
```

Set up the connection to Google Trends:

```python
connection = GoogleConnection()
```

## Single query for daily data

```python
query_single = {'key': 'Apple', 'geo': ''}
start_single = dt.datetime(year=2010, month=1, day=1)
end_single = dt.datetime(year=2019, month=9, day=30)
```

```python
series_single = SVSeries.univariate(connection=connection, query=query_single,
                                   start=start_single, end=end_single, granularity='DAY',
                                   category=CategoryCodes.COMPUTERS_ELECTRONICS)
```

```python
data_single = series_single.get_data()
```

```python
data_single.plot()
```

```python
series_single.request_structure
```

## Multiple queries for monthly data

```python
query_multi = [{'key': 'Mercedes', 'geo': 'US'},
               {'key': 'Porsche', 'geo': 'US'},
               {'key': 'BMW', 'geo': 'US'},
               {'key': 'Lexus', 'geo': 'US'},
               {'key': 'Audi', 'geo': 'US'}]
start_multi = dt.datetime(year=2004, month=1, day=1)
end_multi = dt.datetime(year=2019, month=6, day=1)
```

```python
series_multi = SVSeries.multivariate(connection=connection, queries=query_multi,
                                   start=start_multi, end=end_multi, granularity='MONTH',
                                   category=CategoryCodes.AUTOS_VEHICLES)
```

```python
data_multi = series_multi.get_data()
```

```python
data_multi.plot()
```

```python
series_multi.request_structure
```
