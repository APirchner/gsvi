```python
import datetime as dt
import pandas as pd

import matplotlib.pyplot as plt

%matplotlib inline
plt.rcParams["figure.figsize"] = [15,7.5]

import os
import sys
sys.path.insert(0, os.path.abspath('../'))

from gsvi.connection import GoogleConnection
from gsvi.related import RelatedQueries
from gsvi.catcodes import CategoryCodes
```

Set up the connection to Google Trends:

```python
connection = GoogleConnection()
```

## Related queries for a single query

Specify the query:

```python
query_single = {'key': 'Apple', 'geo': ''}
start_single = dt.datetime(year=2019, month=1, day=1)
end_single = dt.datetime(year=2019, month=12, day=31)
```

Set up the container and fetch data:

```python
related_single = RelatedQueries.single(connection=connection, query=query_single,
                                       start=start_single, end=end_single)
```

```python
related_single.get_data()
```

The data consists of the top queries (maximum is set to 100 by Google) and rising queries (values in %-increase compared to the previous period).

```python
related_single.data[query_single['key']]['top'].plot.bar(x='query', y='value')
```

```python
related_single.data[query_single['key']]['rising'].plot.bar(x='query', y='value')
```

## Related queries for multiple queries

```python
query_multi = [{'key': 'Mercedes Benz', 'geo': 'US'},
               {'key': 'Porsche', 'geo': 'US'},
               {'key': 'BMW', 'geo': 'US'},
               {'key': 'Lexus', 'geo': 'US'},
               {'key': 'Audi', 'geo': 'US'}]
start_multi = dt.datetime(year=2019, month=12, day=1)
end_multi = dt.datetime(year=2019, month=12, day=21)
```

```python
related_multi = RelatedQueries.multiple(connection=connection,
                                        queries=query_multi, start=start_multi, end=end_multi)
```

```python
related_multi.get_data()
```
